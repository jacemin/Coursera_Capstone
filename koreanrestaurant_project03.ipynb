```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns01 = ['name', 'categories'] + [col for col in dataframe01.columns if col.startswith('location.')] + ['id']
dataframe_filtered01 = dataframe01.loc[:, filtered_columns01]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list01 = row['categories']
    except:
        categories_list01 = row['venue.categories']
        
    if len(categories_list01) == 0:
        return None
    else:
        return categories_list01[0]['name']

# filter the category for each row
dataframe_filtered01['categories'] = dataframe_filtered01.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered01.columns = [column.split('.')[-1] for column in dataframe_filtered01.columns]

dataframe_filtered01
```

```python
dataframe_filtered01.name
```

```python
venues_map01 = folium.Map(location=[latitude01, longitude01], zoom_start=16)

folium.CircleMarker(
    [latitude01, longitude01],
    radius=10,
    color='red',
    popup='Broadway',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map01)

for lat, lng, label in zip(dataframe_filtered01.lat, dataframe_filtered01.lng, dataframe_filtered01.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map01)

venues_map01
```
